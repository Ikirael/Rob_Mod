```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon, poisson, binom
```

```python
def show_func(title, x, y, y_dist):    
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.plot(x, y, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title}\nПлотность')
    plt.grid()
    
    plt.subplot(1, 2, 2)
    plt.plot(x, y_dist, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title}\nФункция распределения')
    plt.grid()
    
    plt.show()
```

```python
x = np.linspace(-6, 20, 1000)
y = uniform.pdf(x, loc=1, scale=5)
y_dist = uniform.cdf(x, loc=1, scale=5)
show_func('Равномерное распределение', x, y, y_dist)
```

```python
x = np.linspace(-6, 20, 1000)
y = norm.pdf(x, loc=1, scale=5)
y_dist = norm.cdf(x, loc=1, scale=5)
show_func('Нормальное распределение', x, y, y_dist)
```

```python
x = np.linspace(-6, 20, 1000)
y = expon.pdf(x, loc=1, scale=5)
y_dist = expon.cdf(x, loc=1, scale=5)
show_func('Экспоненциальное" распределение', x, y, y_dist)
```

```python
x = np.linspace(-6, 20, 1000)
y = poisson.pmf(x, mu=0.8)
y_dist = poisson.cdf(x, mu=0.8)
show_func('Распределение Пуассона', x, y, y_dist)
```

```python
x = np.linspace(-6, 20, 1000)
y = binom.pmf(x, n=10, p=0.6)
y_dist = binom.cdf(x, n=10, p=0.6)
show_func('Биномиальное распределение', x, y, y_dist)
```
